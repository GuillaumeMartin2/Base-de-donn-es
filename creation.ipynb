```python
%%sql
drop table if exists g20marti.Emoji;
drop table if exists g20marti.Message;
drop table if exists g20marti.Ami;
drop table if exists g20marti.Demande_Ami;
drop table if exists g20marti.Membre;
drop table if exists g20marti.Permission;
drop table if exists g20marti.Droit;
drop table if exists g20marti.Autorisation;
drop table if exists g20marti.Role;
drop table if exists g20marti.Channel;
drop table if exists g20marti.Serveur;


--Creation de la base de données;

--Création table Serveur;

CREATE TABLE g20marti.Serveur
(   
id          serial primary key NOT NULL,
nom         varchar(64) NOT NULL,
description varchar(64),
date_creation date ,
nom_createur  varchar(64)
      
);

--Création table Channel

CREATE TABLE g20marti.Channel
(   
    id          serial primary key NOT NULL,
    nom         varchar(64),
    typeCanal  varchar(64),
    date_creation  date,
    nom_createur  varchar(64),
    id_serveur int ,
    foreign key(id_serveur) references g20marti.Serveur(id)
      
);

--Creation table Role

CREATE TABLE g20marti.Role 
( id serial primary key NOT NULL,
 nom varchar(64),
 nom_Createur varchar(64),
 Avatar varchar(64),
 id_serveur int,
 foreign key(id_serveur) references g20marti.Serveur(id)
);

--Creation table Membre

CREATE TABLE g20marti.Membre
(   
    id      serial primary key,
    Nom     varchar(64) NOT NULL,
    Pseudo varchar(64),
    mail varchar(64),
    Age int,
    Centre_d_interets varchar(250),
    id_role int,
    id_serveur int,
    foreign key(id_role) references g20marti.Role(id),
    foreign key(id_serveur) references g20marti.Serveur(id)
      
);


--Creation table Message

CREATE TABLE g20marti.Message
(   
    id          serial primary key NOT NULL,
    contenu      varchar(256) NOT NULL,
    date_envoi date,
    heure_envoi  varchar(64) ,
    lieu_envoi varchar(64),
    id_membre int ,
    id_message_ref int,
    id_channel int,
    foreign key(id_membre) references g20marti.Membre(id),
    foreign key(id_message_ref) references g20marti.Message(id),
    foreign key(id_channel) references g20marti.Channel(id)
      
);

--Creation table Emoji

CREATE TABLE g20marti.Emoji
(   
    id      serial primary key NOT NULL,
    nom     varchar(64),
    date_envoi date,
    heure_envoi  varchar(64),
    image varchar(8) ,
    id_membre int,
    id_message_ref int,
    foreign key(id_membre) references g20marti.Membre(id),
    foreign key(id_message_ref) references g20marti.Message(id)
      
);

--Creation table Ami

CREATE TABLE g20marti.Ami
(ref_ami_1 int,
 ref_ami_2 int,
 foreign key(ref_ami_1) references g20marti.Membre(id),
 foreign key(ref_ami_2) references g20marti.Membre(id),
 primary key(ref_ami_1, ref_ami_2)
);

--Creation table Demande_Ami

CREATE TABLE g20marti.Demande_Ami
(
 ref_demande_ami_1 int,
 ref_demande_ami_2 int,
 foreign key(ref_demande_ami_1) references g20marti.Membre(id),
 foreign key(ref_demande_ami_2) references g20marti.Membre(id),
 primary key(ref_demande_ami_1, ref_demande_ami_2)
);

--Creation Autorisation

CREATE TABLE g20marti.Autorisation
(
    ref_channel int,
    ref_role int,
    foreign key(ref_channel) references g20marti.Channel(id),
    foreign key(ref_role) references g20marti.Role(id),
primary key (ref_channel, ref_role));

--Creation Droit

CREATE TABLE g20marti.Droit
(id serial primary key NOT NULL,
 nom varchar(64) NOT NULL,
 Description varchar(128),
 Hierarchie int,
 Date_implementation date
 
);

--Creation Permission

CREATE TABLE g20marti.Permission
(
 id_role int,
 foreign key(id_role) references g20marti.Role(id),
 id_droit int,
 foreign key(id_droit) references g20marti.Droit(id),
 primary key(id_role, id_droit) 
);


--Remplissage 

--Serveur
insert into g20marti.Serveur values (DEFAULT,'IMT Atlantique','Serveur de l IMT Atlantique',to_date('17/05/2020','dd/mm/yyyy'),'Paul Friedel');
insert into g20marti.Serveur values (DEFAULT,'Guillaume','Serveur Personnel',to_date('01/01/2020','dd/mm/yyyy'),'Guillaume Martin');
insert into g20marti.Serveur values (DEFAULT,'Groupe B2 4/4','Serveur du Groupe B2',to_date('18/08/2020','dd/mm/yyyy'),'Guillaume Martin');
insert into g20marti.Serveur values (DEFAULT,'BDS','Serveur du Bureau des Sports ',to_date('10/03/2021','dd/mm/yyyy'),'Gautier Feuillet');
insert into g20marti.Serveur values (DEFAULT,'E-Sport','Serveur du club E-Sport de l IMT',to_date('10/01/2021','dd/mm/yyyy'),'Flavien Deromance');


--Channel
--Serveur IMT
insert into g20marti.Channel values (DEFAULT,'Admin','Vocal',to_date('01/09/2020','dd/mm/yyyy'),'Paul Friedel',1);
insert into g20marti.Channel values (DEFAULT,'Informations ','Textuel',to_date('01/09/2020','dd/mm/yyyy'),'Paul Friedel',1);
insert into g20marti.Channel values (DEFAULT,'Informations  ','Vocal',to_date('01/09/2020','dd/mm/yyyy'),'Paul Friedel',1);
insert into g20marti.Channel values (DEFAULT,'Questions ','Vocal',to_date('01/09/2020','dd/mm/yyyy'),'Paul Friedel',1);
insert into g20marti.Channel values (DEFAULT,'Questions ','Textuel',to_date('01/09/2020','dd/mm/yyyy'),'Paul Friedel',1);

--Serveur Perso
insert into g20marti.Channel values (DEFAULT,'Admin','Vocal',to_date('01/01/2020','dd/mm/yyyy'),'Guillaume Martin',2);
insert into g20marti.Channel values (DEFAULT,'Discussion ','Textuel',to_date('01/01/2020','dd/mm/yyyy'),'Guillaume Martin',2);
insert into g20marti.Channel values (DEFAULT,'Discussion','Vocal',to_date('01/01/2020','dd/mm/yyyy'),'Guillaume Martin',2);
insert into g20marti.Channel values (DEFAULT,'Projets ','Textuel',to_date('01/01/2020','dd/mm/yyyy'),'Guillaume Martin',2);
insert into g20marti.Channel values (DEFAULT,'Projets ','Vocal',to_date('01/01/2020','dd/mm/yyyy'),'Guillaume Martin',2);

--Serveur B2
insert into g20marti.Channel values (DEFAULT,'Admin','Vocal',to_date('05/09/2020','dd/mm/yyyy'),'Guillaume Martin',3);
insert into g20marti.Channel values (DEFAULT,'SER ','Textuel',to_date('06/09/2020','dd/mm/yyyy'),'Guillaume Martin',3);
insert into g20marti.Channel values (DEFAULT,'SER ','Vocal',to_date('05/09/2020','dd/mm/yyyy'),'Guillaume Martin',3);
insert into g20marti.Channel values (DEFAULT,'BD-IHM','Vocal',to_date('07/09/2020','dd/mm/yyyy'),'Flavien Deromance',3);
insert into g20marti.Channel values (DEFAULT,'BD-IHM ','Textuel',to_date('08/09/2020','dd/mm/yyyy'),'Flavien Deromance',3);

--Serveur BDS
insert into g20marti.Channel values (DEFAULT,'Admin','Vocal',to_date('11/03/2021','dd/mm/yyyy'),'Gautier Feuillet',4);
insert into g20marti.Channel values (DEFAULT,'Projets ','Textuel',to_date('11/03/2021','dd/mm/yyyy'),'Gautier Feuillet',4);
insert into g20marti.Channel values (DEFAULT,'Projets ','Vocal',to_date('11/03/2021','dd/mm/yyyy'),'FLorian Nycek',4);
insert into g20marti.Channel values (DEFAULT,'Evenements ','Textuel',to_date('11/03/2021','dd/mm/yyyy'),'Gautier Feuillet',4);
insert into g20marti.Channel values (DEFAULT,'Evenements ','Vocal',to_date('11/03/2021','dd/mm/yyyy'),'Gautier Feuillet',4);
insert into g20marti.Channel values (DEFAULT,'Informations ','Textuel',to_date('12/03/2021','dd/mm/yyyy'),'Gautier Feuillet',4);
insert into g20marti.Channel values (DEFAULT,'Relations exterieures ','Textuel',to_date('13/03/2021','dd/mm/yyyy'),'Rayan Ben Taleb',4);
insert into g20marti.Channel values (DEFAULT,'Communication ','Textuel',to_date('14/03/2021','dd/mm/yyyy'),'Germain Morilhat',4);
insert into g20marti.Channel values (DEFAULT,'Evenements ','Textuel',to_date('11/03/2021','dd/mm/yyyy'),'Gautier Feuillet',4);

--Serveur Esport
insert into g20marti.Channel values (DEFAULT,'Admin ','Vocal',to_date('19/03/2021','dd/mm/yyyy'),'Flavien Deromance',5);
insert into g20marti.Channel values (DEFAULT,'Evenements club  ','Textuel',to_date('19/03/2021','dd/mm/yyyy'),'Flavien Deromance',5);
insert into g20marti.Channel values (DEFAULT,'Evenements club  ','Vocal',to_date('19/03/2021','dd/mm/yyyy'),'Flavien Deromance',5);
insert into g20marti.Channel values (DEFAULT,'Rocket League ','Audio',to_date('19/03/2021','dd/mm/yyyy'),'Bilal Rahou',5);
insert into g20marti.Channel values (DEFAULT,'League of Legends  ','Audio',to_date('22/03/2021','dd/mm/yyyy'),'Guilhem Vermorel',5);


--Role

--Role Serveur Ecole

insert into g20marti.Role values (DEFAULT,'Administrateur','Paul Friedel','Image2 ',1);
insert into g20marti.Role values (DEFAULT,'Eleve','Paul Friedel','Image1 ',1);
insert into g20marti.Role values (DEFAULT,'Professeur','Paul Friedel','Image ',1);

-- Role Serveur Personnel 
insert into g20marti.Role values (DEFAULT,'Administrateur','Guillaume Martin','Image ',2);
insert into g20marti.Role values (DEFAULT,'Invite','Guillaume Martin','Image ',2);

-- Role Serveur B2
insert into g20marti.Role values (DEFAULT,'Administrateur','Guillaume Martin','Image ',3);
insert into g20marti.Role values (DEFAULT,'Eleve','Guillaume Martin','Image ',3);
insert into g20marti.Role values (DEFAULT,'Professeur','Guillaume Martin','Image ',3);
insert into g20marti.Role values (DEFAULT,'Reponsable classe','Guillaume Martin','Image ',3);

--Role Serveur BDS 
insert into g20marti.Role values (DEFAULT,'Administrateur','Gautier Feuillet','Image ',4);
insert into g20marti.Role values (DEFAULT,'Bureau Restreint','Guillaume Martin','Image ',4);
insert into g20marti.Role values (DEFAULT,'Membre simple','Guillaume Martin','Image ',4);

-- Role Serveur Esport
insert into g20marti.Role values (DEFAULT,'Administrateur','Flavien Deromance','Image ',5);
insert into g20marti.Role values (DEFAULT,'Membre du club','Flavien Deromance ','Image ',5);
insert into g20marti.Role values (DEFAULT,'Bureau Restreint','Flavien Deromance ','Image ',5);


--Membre

--Serveur IMT 
insert into g20marti.Membre values (DEFAULT,'Paul Friedel','PFriedel','paul.friedel@imt-atlantique.net',60,'Littérature, Voile',1,1);
insert into g20marti.Membre values (DEFAULT,'Anne Beauval','Abeauval','anne.beauval@imt-atlantique.net',40,'Course à Pied',1,1);
insert into g20marti.Membre values (DEFAULT,'Guillaume Martin','Gmartin','guillaume.martin@imt-atlantique.net',21,'Football, Peche, Jeux Videos',2,1);
insert into g20marti.Membre values (DEFAULT,'Sylvie Kerouedan','SKerouedan','sylvie.kerouedan@imt-atlantique.net',40,'Voile',3,1);

--Serveur Personnel
insert into g20marti.Membre values (DEFAULT,'Guillaume Martin','Gmartin','guillaume.martin@imt-atalantique.net',21,'Football, Peche, Jeux Videos',4,2);
insert into g20marti.Membre values (DEFAULT,'Flavien Deromance','FDeromance','flavien.deromance@imt-atlantique.net',20,'Football,Jeux Videos,Natation ',5,2);
insert into g20marti.Membre values (DEFAULT,'Germain Morilhat','Gmorilhat','germain.morilhat@imt-atalantique.net',20,'Harry Potter,Basketball',5,2);

--Serveur B2
insert into g20marti.Membre values (DEFAULT,'Guillaume Martin','Gmartin','guillaume.martin@imt-atlantique.net',21,'Football, Peche, Jeux Videos',6,3);
insert into g20marti.Membre values (DEFAULT,'Flavien Deromance','Fderomance','flavien.deromance@imt-atlantique.net',20,'Football,Jeux Videos,Natation ',7,3);
insert into g20marti.Membre values (DEFAULT,'Carlos Granero','Ggranero','carlos.granero@imt-atlantique.net',28,'Football',8,3);
insert into g20marti.Membre values (DEFAULT,'Camille Arnold','CArnold','camille.arnold@imt-atlantique.net',20,'Ecologie',9,3);

--Serveur BDS
insert into g20marti.Membre values (DEFAULT,'Gautier Feuillet','GFeuillet','gautier.feuillet@imt-atlantique.net',21,'Judo',10,4);
insert into g20marti.Membre values (DEFAULT,'Florian Nycek','FNycek','florian.nycek@imt-atlantique.net',21,'League of Legends',11,4);
insert into g20marti.Membre values (DEFAULT,'FLorian Gauvrit','FGauvrit','florian.gauvrit@imt-atlantique.net',20,'Olympique Lyonnais',12,4);

--Serveur Esport
insert into g20marti.Membre values (DEFAULT,'Flavien Deromance','Fderomance','flavien.deromance@imt-atlantique.net',20,'Football,Jeux Videos,Natation',13,5);
insert into g20marti.Membre values (DEFAULT,'Jules Saidane','Jsaidane','jules.saidane@imt-atlantique.net',20,'Rocket League',14,5);
insert into g20marti.Membre values (DEFAULT,'Wyatt Marin','Wmarin','wyatt.marin@imt-atlantique.net',20,'FIFA, Football',15,5);


--Message

--Serveur IMT 
insert into g20marti.Message values (DEFAULT,'Bonjour, Bienvenue',to_date('02/09/2020','dd/mm/yyyy'),'18:00:03','Plouzané',1,NULL,2);
insert into g20marti.Message values (DEFAULT,'Bonjour Monsieur',to_date('02/09/2020','dd/mm/yyyy'),'18:00:47','Paris',3,1,2);

insert into g20marti.Message values (DEFAULT,'Discussion vocale',to_date('02/09/2020','dd/mm/yyyy'),'18:03:47','Plouzané',1,NULL,1);
insert into g20marti.Message values (DEFAULT,'Réponse vocale',to_date('02/09/2020','dd/mm/yyyy'),'18:03:58','Plouzané',2,3,1);
insert into g20marti.Message values (DEFAULT,'Les cours commencent bien Lundi?',to_date('02/09/2020','dd/mm/yyy'),'18:17:18','Plouzané',3,NULL,5);
insert into g20marti.Message values (DEFAULT,'Oui',to_date('02/09/2020','dd/mm/yyyy'),'18:18:18','Paris',1,5,5);
insert into g20marti.Message values (DEFAULT,'D autres questions ? ',to_date('02/09/2020','dd/mm/yyyy'),'18:21:23','Plouzané',1,NULL,5);
insert into g20marti.Message values (DEFAULT,'Non merci ',to_date('02/09/2020','dd/mm/yyyy'),'18:22:18','Paris',3,NULL,5);

--Serveur Personnel
insert into g20marti.Message values (DEFAULT,'Salut, ça va ? ',to_date('17/05/2021','dd/mm/yyyy'),'18:00:00','Plouzané',5,NULL,7);
insert into g20marti.Message values (DEFAULT,'Oui et toi ?',to_date('17/05/2021','dd/mm/yyyy'),'18:00:56','Plouzané',6,9,7);
insert into g20marti.Message values (DEFAULT,'Très bien ',to_date('17/05/2021','dd/mm/yyyy'),'18:03:00','Plouzané',5,10,7);
insert into g20marti.Message values (DEFAULT,'Se renseigner sur le Portugal',to_date('18/05/2021','dd/mm/yyyy'),'10:00:00','Plouzané',5,NULL,9);
insert into g20marti.Message values (DEFAULT,'Espagne',to_date('18/05/2021','dd/mm/yyyy'),'10:00:12','Plouzané',7,NULL,9);
insert into g20marti.Message values (DEFAULT,'Finlande',to_date('18/05/2021','dd/mm/yyyy'),'10:03:12','Plouzané',5,NULL,9);


--Emoji

--Serveur IMT 
insert into g20marti.Emoji values (DEFAULT,'Sourire',to_date('30/05/2021','dd/mm/yyyy'),'18:30:00','Image',3,1);
insert into g20marti.Emoji values (DEFAULT,'Pouce Haut',to_date('30/05/2021','dd/mm/yyyy'),'18:31:00','Image1',3,3);

--Serveur Personnel
insert into g20marti.Emoji values (DEFAULT,'Pouce Haut',to_date('30/05/2021','dd/mm/yyyy'),'18:30:00','Image1',5,11);


--Ami

--Creation des amitiés

insert into g20marti.Ami values (1,2);
insert into g20marti.Ami values (2,4);
insert into g20marti.Ami values (1,4);
insert into g20marti.Ami values (5,6);
insert into g20marti.Ami values (5,7);
insert into g20marti.Ami values (6,7);
insert into g20marti.Ami values (8,9);
insert into g20marti.Ami values (8,11);
insert into g20marti.Ami values (9,11);

--Creation des Demande_AMI
insert into g20marti.Demande_Ami values (3,1);
insert into g20marti.Demande_Ami values (3,2);

--Autorisation

--Serveur IMT

--Admin serveur IMT
insert into g20marti.Autorisation values(1,1);
insert into g20marti.Autorisation values(2,1);
insert into g20marti.Autorisation values(3,1);
insert into g20marti.Autorisation values(4,1);
insert into g20marti.Autorisation values(5,1);

-- eleve IMT
insert into g20marti.Autorisation values(2,2);
insert into g20marti.Autorisation values(3,2);
insert into g20marti.Autorisation values(4,2);
insert into g20marti.Autorisation values(5,2);

-- Prof IMT
insert into g20marti.Autorisation values(2,3);
insert into g20marti.Autorisation values(3,3);
insert into g20marti.Autorisation values(4,3);
insert into g20marti.Autorisation values(5,3);

-- Serveur Perso
-- Admin
insert into g20marti.Autorisation values(6,4);
insert into g20marti.Autorisation values(7,4);
insert into g20marti.Autorisation values(8,4);
insert into g20marti.Autorisation values(9,4);
insert into g20marti.Autorisation values(10,4);

-- Invité
insert into g20marti.Autorisation values(7,5);
insert into g20marti.Autorisation values(8,5);
insert into g20marti.Autorisation values(9,5);
insert into g20marti.Autorisation values(10,5);

-- Droit
insert into g20marti.Droit values (DEFAULT,'Bannir','Exclure un utilisateur du serveur',1,to_date('02/09/2020','dd/mm/yyyy'));
insert into g20marti.Droit values (DEFAULT,'Supprimer un message','Supprimer un message incorrect',3,to_date('02/09/2020','dd/mm/yyyy'));
insert into g20marti.Droit values (DEFAULT,'Créér','Créer un nouveau channel dans le serveur',2,to_date('02/09/2020','dd/mm/yyyy'));
insert into g20marti.Droit values (DEFAULT,'Modifier serveur','Modifier des informations du serveur',2,to_date('02/09/2020','dd/mm/yyyy'));


-- Permission 
-- Serveur IMT
insert into g20marti.Permission values (1,1);
insert into g20marti.Permission values (1,2);
insert into g20marti.Permission values (1,3);
insert into g20marti.Permission values (1,4);


-- Serveur Perso
insert into g20marti.Permission values (3,1);
insert into g20marti.Permission values (3,2);
insert into g20marti.Permission values (3,3);
insert into g20marti.Permission values (3,4);

insert into g20marti.Permission values (4,2);
insert into g20marti.Permission values (4,4);


```
